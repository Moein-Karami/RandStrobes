```python
# !pip3 install pandas
import pandas as pd
# !pip3 install matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('final_result.csv')

print(df.to_string()) 

window_sizes = df["Window_Size"].tolist()[:8]

mamd_time_executions = [x / 1000000 for x in df["Execution_Time"].tolist()[:8]]
mamd_ehits = df["final_seed_hash_value_E-hits"].tolist()[:8]

cc_time_executions = df["Execution_Time"].tolist()[8:16]
cc_ehits = df["final_seed_hash_value_E-hits"].tolist()[8:16]

kmer_time_executions = df["Execution_Time"].tolist()[16:] * 8
kmer_ehits = df["final_seed_hash_value_E-hits"].tolist()[16:] * 8
```

```python
plt.plot(window_sizes, mamd_ehits, '-', label = 'MAMD')
plt.plot(window_sizes, cc_ehits, '-.', label = "CC")
plt.plot(window_sizes, kmer_ehits, label = "K-mer")

plt.xlabel("Window Size")
plt.ylabel("E-hits")
plt.title('E-hits of final seed hash value')
plt.legend()
plt.savefig('all_ehits.png')
plt.show()
```

```python
plt.plot(window_sizes, mamd_time_executions, '-', label = 'MAMD')
plt.plot(window_sizes, cc_time_executions, '-.', label = "CC")
plt.plot(window_sizes, kmer_time_executions, label = "K-mer")

plt.yscale('log')
plt.xlabel("Window Size")
plt.ylabel("E-hits")
plt.title('Execution time')
plt.legend()
plt.savefig('all_execution-time.png')
plt.show()
```
